### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter plots are the graphs that present the relationship between two variables in a data-set. It represents data points on a two-dimensional plane or on a Cartesian system. The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis. These plots are often called scatter graphs or scatter diagrams.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y, c ="blue")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("X-Y Plot")
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

subplot() function adds subplot to a current figure at the specified grid position.It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. So to create multiple plots you will need several lines of code with the subplot() function.

```python
import numpy as np
import matplotlib.pyplot as plt

#For line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

#For line 2: 
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

#For line 3: 
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

#For line 4: 
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
             
plt.figure()
 
# adding first subplot
plt.subplot(221)
plt.plot(x1, y1, color="orange", marker="*")
 
# adding second subplot
plt.subplot(222)
plt.plot(x2, y2, color="yellow", marker="*")

# adding third subplot
plt.subplot(223)
plt.plot(x3, y3, color="orange", marker="*")

# adding third subplot
plt.subplot(224)
plt.plot(x4, y4, color="yellow", marker="*")
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.

Bar charts should be used when you are showing segments of information. Vertical bar charts are useful to compare different categorical or discrete variables, such as age groups, classes, schools, etc., as long as there are not too many categories to compare. They are also very useful for time series data.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.xlabel("Companies name")
plt.ylabel("Profit")
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

It is a type of chart that depicts a group of numerical data through their quartiles. It is a simple way to visualize the shape of our data. It makes comparing characteristics of data between categories very easy.

Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# Pandas dataframe
data = pd.DataFrame({"Box1": box1, "Box2": box2})

# Plot the dataframe
ax = data[['Box1', 'Box2']].plot(kind='box', title='boxplot')

# Display the plot
plt.show()
```
